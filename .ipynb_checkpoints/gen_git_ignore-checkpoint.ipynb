```python
#Generate gitignore with files greater 100 MB
import os
import sys
from glob import glob
```

```python
#Files to add to gitignore
ignore_list = ['gen_git_ignore.ipynb', 
                'pred/script/pafnucy/results/',
                'si_data/script/',
                'si_data/pdb_with_fad',
                'si_data/top_dock_pose']
```

```python
files = [file for file in glob("**/*", recursive = True) if os.path.isfile(file)]
files
```

```python
for file in files:
    file_size = os.path.getsize(file)
    if file_size > 1e+8:
        print(file, file_size / 1e+6, 'MB')
        ignore_list.append(file)
```

```python
print(ignore_list)
```

```python
#Generate gitignore
with open('.gitignore', 'w') as writer:
    for x in ignore_list:
        writer.write(x + '\n')
```
